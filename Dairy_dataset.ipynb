```python
m=df.groupby(df['Product Name'],as_index=False)['Quantity Sold (liters/kg)'].sum().sort_values(by='Quantity Sold (liters/kg)', ascending=False)

d=sns.barplot(x='Product Name', y='Quantity Sold (liters/kg)',data=m)
for bars in d.containers:
    d.bar_label(bars)
```

```python
w=sns.barplot(x=df['Sales Channel'],y=df['Approx. Total Revenue(INR)'])
for bars in w.containers:
    w.bar_label(bars)
```

```python
y=df.groupby(df['Sales Channel'],as_index=False)['Quantity Sold (liters/kg)'].sum().sort_values(by='Quantity Sold (liters/kg)',ascending=False)

sns.set(rc={'figure.figsize':(15,5)})
g=sns.barplot(x='Sales Channel',y='Quantity Sold (liters/kg)',data=y)

for bars in g.containers:
    g.bar_label(bars)
```

```python
x=df.groupby(df['Storage Condition'],as_index=False)['Approx. Total Revenue(INR)'].sum().sort_values(by='Approx. Total Revenue(INR)',ascending=False)

h=sns.barplot(x='Storage Condition',y='Approx. Total Revenue(INR)',data=x)

for bars in h.containers:
    h.bar_label(bars)
```

```python
matplotlib.rcParams['figure.figsize']=(20,5)
x=df.groupby(df['Location'],as_index=False)['Total Land Area (acres)'].sum().sort_values(by='Total Land Area (acres)',ascending=False)

h=sns.barplot(x='Location',y='Total Land Area (acres)',data=x)

for bars in h.containers:
    h.bar_label(bars)
```

```python
matplotlib.rcParams['figure.figsize']=(20,5)
x=df.groupby(df['Product Name'],as_index=False)['Quantity Sold (liters/kg)'].sum().sort_values(by='Quantity Sold (liters/kg)',ascending=False)

h=sns.barplot(x='Product Name',y='Quantity Sold (liters/kg)',data=x)

for bars in h.containers:
    h.bar_label(bars)
```
